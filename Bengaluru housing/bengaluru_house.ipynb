```python
df['total_sqft'].value_counts()
df['total_sqft'].unique()
sqft=[]
for d in df['total_sqft']:
    if '-' in d:
        k=d.split(' - ')
        sqft.append(float(k[0])+float(k[1])/2)
    elif d.isnumeric():
        sqft.append(d)
        
        
    else:
        sqft.append(None)
df['total_sqft']=sqft
df['total_sqft']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
lr=DecisionTreeRegressor()
lr.fit(x_train,y_train)
ypredict=lr.predict([[3,40,419,1056,2,1,2]])
ypredict=lr.predict(x_test)

# yactual=np.array([3907000])
# yactual
ypredict
```

```python
lr.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,ypredict))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,ypredict))
```
